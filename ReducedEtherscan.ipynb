```python
#Pull in the operator data and list of missing wallets
walletData = pd.read_pickle('operatorData.pkl')
missingWallets = pd.read_csv('missingWallets.csv')
missing = missingWallets.wallet.unique()
wallets = walletData[~walletData.wallet.isin(missing)]
len(wallets)
```

```python
walletList = wallets.wallet.unique()
```

```python
#return only the wallet IDs of transaction partners who are known Storj node operators
def associateFinder(row):
    return list(set([x for x in row if x in walletList]))
```

```python
#For each operator store a list of the wallets they've sent tokens to
justTo = full[full.to.isin(walletList)]
justTo['wallet'] = justTo.to
justTo
```

```python
#for each operator store a list of all wallets they've received tokens from
justFrom = full[full['from'].isin(walletList)]
justFrom['wallet'] = justFrom['from']
justFrom
```

```python
#txs = justTo.merge(justFrom, left_on='hash',right_on='hash',how='outer')
txs = pd.concat([justTo, justFrom]).reset_index(drop=True)
txs
```
